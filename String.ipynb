```python
text = "Julide"

def reverse_string(text):
    num = len(text)          # "Julide" → 6 karakter
    reserved_text = ""       # ters string için boş başlangıç
    while num > 0:           # num sıfır olana kadar döngü
        num = num - 1        # her adımda 1 azalt
        reserved_text += text[num]   # sondan başa karakter ekle
    return reserved_text

print(reverse_string(text))
```
